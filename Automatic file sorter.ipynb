```python
import os, shutil

path = '/Users/Academics/Downloads/Automatic file sorter/'

folder_names = ['CSV Files', 'Text Files', 'Image Files', 'Excel Files', 'PDF Files','Msword Files', 'Video Files', 'Unknown Files', 'Powerpoint Files']

for folder in folder_names:
    if not os.path.exists(path + folder):
        os.makedirs(path + folder)

file_names = os.listdir(path)

for file in file_names:
    if ".csv" in file.lower() and not os.path.exists(path + 'CSV Files/' + file):
        shutil.move(path + file ,path + 'CSV Files/' + file)
    elif ".txt" in file.lower() and not os.path.exists(path + 'Text Files/' + file):
        shutil.move(path + file ,path + 'Text Files/' + file)
    elif ".xlsx" in file.lower() and not os.path.exists(path + 'Excel Files/' + file):
        shutil.move(path + file ,path + 'Excel Files/' + file)
    elif ".pdf" in file.lower() and not os.path.exists(path + 'PDF Files/' + file):
        shutil.move(path + file ,path + 'PDF Files/' + file)
    elif ('.jpeg' in file.lower() or '.jpg' in file.lower() or '.png' in file.lower() or '.gif' in file.lower() or '.bmp' in file.lower() or '.tiff' in file.lower() or '.tif' in file.lower() or '.webp' in file.lower() or '.svg' in file.lower() or '.heif' in file.lower() or '.heic' in file.lower() or '.ico' in file.lower()) and not os.path.exists(path + 'Image Files/' + file):
        shutil.move(path + file, path + 'Image Files/' + file)
    elif ('.docx' in file.lower() or '.doc' in file.lower()) and not os.path.exists(path + 'Msword Files/' + file):
        shutil.move(path + file, path + 'Msword Files/' + file)
    elif ('.mp4' in file.lower() or '.mov' in file.lower() or '.avi' in file.lower() or '.mkv' in file.lower() or '.wmv' in file.lower() or '.flv' in file.lower() or '.webm' in file.lower() or '.3gp' in file.lower() or '.mpeg' in file.lower() or '.mpg' in file.lower()) and not os.path.exists(path + 'Video Files/' + file):
        shutil.move(path + file, path + 'Video Files/' + file)
    elif ".sql" in file.lower() and not os.path.exists(path + 'MySQL/' + file):
        shutil.move(path + file ,path + 'MySQL/' + file)
    elif ".pptx" in file.lower() and not os.path.exists(path + 'Powerpoint Files/' + file):
        shutil.move(path + file ,path + 'Powerpoint Files/' + file)
    else:
        if os.path.isfile(path + file) and not os.path.exists(path + 'Unknown Files/' + file):
            shutil.move(path + file, path + 'Unknown Files/' + file)
        


```
