# Lets put it all together 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
from datetime import datetime
import time
```

```python
def automated_crypto_pull():
    
    url = 'https://coinmarketcap.com/currencies/bitcoin/'
    
    page = requests.get(url)
    
    soup = BeautifulSoup(page.text, 'html')
    
    tag = soup.find('span', class_ = 'sc-65e7f566-0 lsTl')
    
    crypto_name = tag['title']
    
    crypto_price = soup.find('span', class_ = 'sc-65e7f566-0 WXGwg base-text').text
    
    final_price = crypto_price.replace('$','')
    
    date_time = datetime.now()
    
    dict = {'Crypto Name': crypto_name,
           'Price': final_price,
           'TimeStamp': date_time}
    
    df = pd.DataFrame([dict])
    
    if os.path.exists('/Users/Academics/Downloads/Automatic file sorter/CSV Files/Crypto web puller/Crypto_Automated_Pull.csv'):
        df.to_csv('/Users/Academics/Downloads/Automatic file sorter/CSV Files/Crypto web puller/Crypto_Automated_Pull.csv', mode = 'a', header = False, index = False)
    else:
        df.to_csv('/Users/Academics/Downloads/Automatic file sorter/CSV Files/Crypto web puller/Crypto_Automated_Pull.csv', index = False)

    print(df)

```

```python
while True:
    automated_crypto_pull()
    time.sleep(10)
```
